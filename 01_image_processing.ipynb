## Image Processing

```python
bgr_img = cv2.imread('./porygon.jpg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
print(type(bgr_img)) # np.array
print(bgr_img.shape, bgr_img.dtype) # (600:row(y), 600:column(x), 3:color), np.uint8
print(bgr_img[200, 150])
print(bgr_img[300, 100])
plt.imshow(rgb_img[199:202, 149:152])
```

```python
# 300 x 300 resize
half_img = rgb_img[::2, ::2]
print(half_img.shape)
plt.imshow(half_img)
```

```python
# color -> gray

img_gray = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)

# numpy color -> gray
img_gray = 0.2126 * bgr_img[:, :, 2] + 0.7152 * bgr_img[:, :, 1] + 0.0722 * bgr_img[:, :, 0]
img_gray = img_gray.astype(np.uint8)
print(img_gray.shape)
print(img_gray.dtype)

plt.imshow(img_gray, cmap='gray')
```

```python
img_zero = np.zeros((600, 600), dtype=np.uint8)
img_zero[img_gray > 210] = 255 # 255 == True, 0 == False
img_zero = 255 - img_zero
plt.imshow(img_zero, cmap='gray')
```

```python
img_src = img_zero.copy()

img_erode = cv2.erode(img_src, (10, 10), iterations=2)
# img_erode = cv2.erode(img_erode, (10, 10))

img_dil = cv2.dilate(img_erode, (10, 10), iterations=8)

plt.imshow(img_erode, cmap='gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
print(kernel)

# erode, dilate -> 모폴로지 연산

img_morph = cv2.morphologyEx(img_src, cv2.MORPH_ERODE, kernel, iterations=1)
img_morph = cv2.morphologyEx(img_morph, cv2.MORPH_CLOSE, kernel, iterations=50)

plt.imshow(img_morph, cmap='gray')
```

```python
img_threshold = np.where(img_morph == 255, img_gray, 0)
plt.imshow(img_threshold, cmap='gray')
```
